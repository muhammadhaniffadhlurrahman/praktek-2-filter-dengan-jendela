```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift
import scipy.signal as sig

# Konfigurasi ukuran font untuk plot
FONT_SMALL = 12
FONT_MEDIUM = 14

# Mengatur ukuran font untuk judul dan label sumbu pada plot
plt.rc('axes', titlesize=FONT_MEDIUM)
plt.rc('axes', labelsize=FONT_SMALL)

# Mengatur ukuran font untuk label pada sumbu x dan y pada plot
plt.rc('xtick', labelsize=FONT_SMALL)
plt.rc('ytick', labelsize=FONT_SMALL)

# Mengatur ukuran font untuk legenda pada plot
plt.rc('legend', fontsize=FONT_MEDIUM)

# Mengatur ukuran font untuk judul pada gambar (figure)
plt.rc('figure', titlesize=FONT_MEDIUM)

# Mengaktifkan mode inline untuk matplotlib di Jupyter Notebook
%matplotlib inline
```

```python
# Parameter masukan
N = 256
NFFT = 2048

# Sinyal masukan (panjang: N >> NFFT)
tt = np.linspace(0, 1, N, endpoint=True)
xx = np.cos(2*np.pi*32*tt)

# Menghitung FFT
sft = np.abs(fft(xx, NFFT))
slg = 20*np.log10(sft / np.max(sft))

# Membuat gambar (figure) dengan ukuran dan resolusi tertentu
plt.figure(figsize=(16, 6), dpi=80)

# Plot sinyal periodik
plt.subplot(2,1,1)
plt.title('Sinyal Periodik, N = {}'.format(N))
plt.xlabel('Sampel waktu')
plt.ylabel('Magnitudo')
plt.plot(xx)
plt.xlim([0, N-1])
plt.grid(True)

# Plot spektrum
plt.subplot(2,1,2)
plt.title('Spektrum, NFFT = {}'.format(NFFT))
plt.xlabel('Sampel frekuensi')
plt.ylabel('Level [dB]')
plt.plot(slg)
plt.xlim([0, NFFT//2])
plt.ylim([-40, 0])
plt.grid(True)
plt.tight_layout()
```

Sinyal ini memiliki bentuk yang terulang setiap beberapa sampel, yang mengindikasikan sifat periodiknya. 'N = 256' menunjukkan bahwa data sampel terdiri dari 256 poin.

2. Spektrum, nFFT = 2048: Grafik bagian bawah menunjukkan spektrum dari sinyal di bagian atas, yang biasanya menganalisis komponen frekuensi sinyal. Konsentrasi puncak di sekitar frekuensi rendah menunjukkan bahwa sebagian besar energi sinyal terkonsentrasi di frekuensi rendah, dan menurun seiring bertambahnya frekuensi.

```python
# Jumlah sampel sinyal
N = 65536

# Membuat array waktu
tt = np.linspace(0, 1, N, endpoint=True)

# Membuat sinyal kosinus dengan frekuensi 32 Hz
xx = np.cos(2*np.pi*32*tt)

# Menghitung FFT
sft = np.abs(fft(xx, N))
slg = 20*np.log10(sft / np.max(sft))

# Membuat gambar (figure) dengan ukuran tertentu
plt.figure(figsize=(16, 6))

# Plot sinyal periodik
plt.subplot(2,1,1)
plt.title('Sinyal Periodik, N = {}'.format(N))
plt.xlabel('Sampel waktu')
plt.ylabel('Magnitudo')
plt.plot(xx)
plt.xlim([0, N-1])
plt.grid(True)

# Plot spektrum
plt.subplot(2,1,2)
plt.title('Spektrum, NFFT = {}'.format(N))
plt.xlabel('Sampel frekuensi')
plt.ylabel('Level [dB]')
plt.plot(slg)
plt.xlim([0, N//256])
plt.ylim([-60, 0])
plt.grid(True)
plt.tight_layout()
```

Gambar diatas menunjukkan dua plot yang berbeda terkait dengan pemrosesan sinyal atau analisis frekuensi. Berikut adalah penjelasan untuk masing-masing plot:

2. Plot Kedua (bawah) - Spektrum, NFFT = 65536:
 - Grafik ini adalah representasi dari spektrum frekuensi sinyal yang ditunjukkan di plot pertama, biasanya dihasilkan menggunakan Fast Fourier Transform (FFT).

 - Sumbu vertikal menunjukkan "Level" dalam dB (decibel), yang mengukur intensitas setiap frekuensi dalam sinyal.
 - Grafik ini menunjukkan hanya satu puncak, yang terdapat di dekat frekuensi yang sangat rendah (dekat dengan 0). Ini mengindikasikan bahwa sinyal asli merupakan sinyal sinusoid murni dengan satu frekuensi dominan, mendekati 0 (atau frekuensi dasar sinyal tersebut).

Kesimpulan: Grafik sinyal periodik dan spektrumnya ini biasanya digunakan dalam teknik dan fisika untuk menganalisis karakteristik frekuensi sinyal dan untuk menentukan seberapa efektif suatu sinyal dalam menyebar frekuensinya. Spektrumnya menunjukkan bahwa sinyal ini hanya memiliki satu frekuensi dominan, yang sesuai dengan penampilan sinusoidal yang sangat teratur pada plot sinyal waktu.

```python
def dft_win(w):
    """
    Fungsi untuk plot fungsi jendela dan spektrum dari jendela.

    Parameters
    ----------
    w : np.array
        vektor jendela masukan
    """

    # Menemukan panjang jendela
    N = len(w)
    # Ukuran FFT
    NFFT = 2**12
    # Menghitung FFT
    W = fftshift(fft(w, NFFT))
    # Normalisasi nilai maksimum FFT dan menambahkan nilai floating-point minimum (kecuali jika pembagian dengan nol)
    W = W / np.amax(W) + np.nextafter(0,1)
    tt = np.linspace(-1, 1, NFFT)

    # Plot fungsi jendela dan spektrumnya
    fig = plt.figure(figsize=(16, 6))

    plt.subplot(2,1,1)
    plt.stem(w, use_line_collection=True, basefmt='C0')
    plt.title('Fungsi Jendela')
    plt.xlabel('Sampel')
    plt.ylabel(r'$w[k]$')
    plt.xlim([0, N-1])
    plt.grid(True)

    plt.subplot(2,1,2)
    plt.plot(tt, 20*np.log10(np.abs(W)))
    plt.title('Spektrum')
    plt.xlabel('Frekuensi')
    plt.ylabel('Level dalam dB')
    plt.axis([-1, 1, -120, 2])
    plt.grid(True)
    plt.tight_layout()
```

```python
# Menggunakan fungsi dft_win untuk plot fungsi jendela dari vektor berisi N elemen dengan nilai satu

N = 64
dft_win(np.ones(N))
```

Gambar diatas merupakan dua buah grafik yang berkaitan dengan pelaksanaan dan representasi sinyal dalam domain waktu dan frekuensi. Mari kita analisa masing-masing grafik:

1. Grafik Pertama (atas): Fungsi Jendela
 - Grafik ini memperlihatkan fungsi jendela, yang biasanya digunakan dalam pemrosesan sinyal untuk mengisolasi segmen tertentu dari sinyal lebih besar.
 - Sumbu horizontal (sumbu X) menunjukkan "Sampel" yang mengindikasikan indeks sampel atau titik data dalam ruang waktu.
 - Sumbu vertikal (sumbu Y) menunjukkan "Amplitudo" dalam persentase (%), yang mengukur intesitas atau besar kecilnya sinyal.
 - Kurva yang terlihat berbentuk lonceng (bell-shaped) yang melonjak naik mencapai puncak dan kemudian turun kembali, mengindikasikan bahwa amplitudo tertinggi terfokus di tengah dan mengecil ke arah tepi.

2. Grafik Kedua (bawah): Spektrum
 - Grafik bawah menunjukkan spektrum frekuensi dari jendela waktu yang ditampilkan di atas.
 - Sumbu horizontal (sumbu X) menunjukkan "Frekuensi", yang mengukur seberapa sering suatu komponen frekuensi terjadi dalam sinyal.
 - Sumbu vertikal (sumbu Y) menunjukkan "Intensitas (db)" dalam logaritma desibel, yang merupakan ukuran logaritmik dari amplitudo.
 - Kurva menunjukkan bahwa energi terpusat pada frekuensi rendah dengan penurunan amplitudo ketika mendekati frekuensi yang lebih tinggi.

```python
# Mendefinisikan panjang jendela (N) dan matriks ww untuk menyimpan berbagai jenis jendela
N = 64

# Membuat matriks 2D (N x 9) untuk menyimpan berbagai jenis jendela
ww = np.zeros((N, 9))

# Mengisi kolom-kolom matriks ww dengan berbagai jenis jendela
ww[:,0] = np.ones(N)  # Jendela rectangular
ww[:,1] = sig.triang(N)  # Jendela Bartlett (triangular)
ww[:,2] = sig.hann(N)  # Jendela Hann (Hanning)
ww[:,3] = sig.hamming(N)  # Jendela Hamming
ww[:,4] = sig.blackman(N)  # Jendela Blackman
ww[:,5] = sig.blackmanharris(N)  # Jendela Blackman-Harris 4-term
ww[:,6] = sig.nuttall(N)  # Jendela Nuttall
ww[:,7] = sig.flattop(N)  # Jendela Flat-top
ww[:,8] = sig.kaiser(N, beta=8)  # Jendela Kaiser dengan beta=8

# Daftar judul untuk tiap jenis jendela
lst_titles = ['Rectangular', 'Bartlett', 'Hann (Hanning)',
              'Hamming', 'Blackman', 'Blackman-Harris 4-term',
              'Nuttall', 'Flat-top window', 'Kaiser (w/ beta=8)'
             ]

# Plot fungsi jendela dan spektrumnya
fig = plt.figure(figsize=(16, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.stem(ww[:,i], use_line_collection=True, basefmt='C0')
    plt.title(lst_titles[i])
    plt.xlabel('Sampel')
    plt.ylabel(r'$w[k]$')
    plt.xlim([0, N-1])
    plt.grid(True)
plt.tight_layout()
```

Gambar diatas menunjukkan berbagai jenis jendela (window functions) yang digunakan dalam pemrosesan sinyal digital. Jendela-jendela ini digunakan untuk memodifikasi sinyal sebelum transformasi Fourier, yang membantu mengurangi kebocoran spektral. Setiap grafik menampilkan tipe jendela yang berbeda, dengan sumbu horizontal yang melambangkan jumlah sampel dan sumbu vertikal yang melambangkan amplitudo. Berikut adalah penjelasan untuk masing-masing jendela yang ditampilkan:

1. Rectangular: Jendela ini memiliki bentuk sederhana di mana semua nilai amplitudo konstan dan kemudian langsung turun menjadi nol. Ini tidak memberikan penghalusan apapun dan bisa menghasilkan kebocoran spektral yang signifikan.

2. Bartlett: Juga dikenal sebagai jendela segitiga, ini secara bertahap meningkat sampai ke titik tengah dan kemudian menurun. Penghalusan ini mengurangi kebocoran spektral tetapi tidak seefektif jendela lainnya.

3. Hann (Hanning): Jendela ini memiliki bentuk parabola dan dikenal baik dalam mengurangi kebocoran spektral dengan penghalusan yang efektif di sisi jendela.

4. Hamming: Mirip dengan jendela Hann tetapi dengan puncak yang sedikit lebih rendah di tepi, memberikan proteksi lebih terhadap kebocoran spektral.

5. Blackman: Ini memiliki bentuk yang lebih halus daripada Hann atau Hamming, menggunakan tiga istilah kosinus untuk mengurangi kebocoran lebih lanjut.

6. Blackman-Harris 4-term: Versi yang lebih canggih dari jendela Blackman, menggunakan empat istilah kosinus, memberikan salah satu penurunan kebocoran spektral terbaik yang tersedia.

7. Nuttall: Mirip dengan Blackman-Harris, tetapi dengan koefisien yang berbeda untuk istilah kosinusnya, memberikan sifat penurunan spektrum yang sangat baik.

8. Flat-Top window: Dirancang untuk memberikan amplitudo yang sangat datar di puncaknya, sangat berguna untuk aplikasi pengukuran amplitudo yang memerlukan akurasi tinggi.

9. Kaiser (w/ beta = 8.6): Jendela yang parameterisasi dengan parameter beta, yang dapat disesuaikan untuk memberikan kompromi antara lebar pita utama dan tingkat penurunan di luar pita.

Setiap jendela memiliki keunggulan dan kelemahan tersendiri, dan pemilihan jendela tertentu sering kali tergantung pada aplikasi khusus yang dihadapi.

```python
NFFT = 2**12  # Panjang FFT (jumlah titik FFT)

# Menghitung FFT dari matriks jendela ww dengan panjang NFFT
WW = fft(ww, NFFT, axis=0)

# Normalisasi FFT agar nilainya dalam rentang yang diharapkan
WW = WW / np.amax(WW, axis=0) + np.nextafter(0, 1)

# Generate array waktu untuk plot spektrum
tt = np.linspace(-1, 1, NFFT)

# Plot fungsi jendela dan spektrumnya
fig2 = plt.figure(figsize=(16, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    # Plot spektrum dari jendela menggunakan FFT yang telah dimodifikasi
    plt.plot(tt, 20*np.log10(np.abs(fftshift(WW[:,i]))))
    plt.title(lst_titles[i])  # Judul plot berdasarkan jenis jendela
    plt.xlabel('Frequency')  # Label sumbu x
    plt.ylabel('Level in dB')  # Label sumbu y
    plt.axis([-1, 1, -120, 2])  # Rentang sumbu x dan y
    plt.grid(True)  # Aktifkan grid pada plot
plt.tight_layout()  # Optimalkan tata letak plot untuk menghindari tumpang tindih
```

Gambar diatas menunjukkan sembilan grafik yang mengilustrasikan spektrum kebocoran dari berbagai jenis jendela (window functions) dalam pemrosesan sinyal. Setiap grafik memberikan visualisasi dari respons frekuensi jendela tersebut, diukur dalam dB terhadap frekuensi. Berikut adalah penjelasan dari setiap jendela yang ditampilkan:

1. Rectangular:
 - Jendela segi empat sederhana tanpa penimbangan, semua titik memiliki bobot yang sama.
 - Mempunyai lebar pita utama yang sempit tetapi tingkat sidelobes yang tinggi, yang menyebabkan kebocoran yang signifikan.

2. Bartlett:

 - Juga dikenal sebagai jendela segitiga.
 - Memiliki bentuk segitiga yang menyebabkan penurunan bobot ke arah ujung jendela, mengurangi kebocoran dibandingkan dengan jendela segi empat.

3. Hann (Hanning):
 - Bentuknya lebih melengkung dan mulus, didesain untuk mengurangi kebocoran lebih lanjut.
 - Berguna dalam mengurangi kebocoran sinyal saat kecilnya pemusatan sinyal adalah lebih penting dari resolusi frekuensi.

4. Hamming:
 - Mirip dengan Hanning tetapi dengan sedikit variasi pada parameter untuk mencapai tingkat sidelobe yang lebih rendah.

5. Blackman:
 - Lebih rumit daripada Hann atau Hamming.
 - Dirancang untuk menurunkan sidelobes lebih lanjut tetapi dengan konsekuensi memperlebar lebar pita utama.

6. Flat-top window:
 - Maksimal menghilangkan kesalahan dalam pengukuran amplitude spektrum, dengan sangat meratakan puncak (flat-top).
 - Sering digunakan dalam aplikasi pengukuran spektrum karena akurasinya yang tinggi pada amplituda.

7. Blackman-Harris:
 - Pengembangan lebih lanjut dari jendela Blackman, menggunakan lebih banyak koefisien untuk memperbaiki penurunan kebocoran.

8. Nuttall:
 - Memiliki kombinasi performa yang baik untuk meminimalkan baik lebar pita utama dan amplitudo sidelobes.

9. Kaiser (I0-Beta=8):
 - Jendela yang dapat diatur keberadaannya berdasarkan nilai beta, dengan flexibilitas untuk menyesuaikan kedalaman sidelobe dan lebar pita utama.
 - Pilihan populer untuk berbagai aplikasi khusus karena fleksibilitasnya dalam pengaturan sesuai kebutuhan spesifik.

Setiap jendela ini memiliki karakteristik tertentu yang membuatnya cocok untuk aplikasi yang berbeda tergantung pada kebutuhan terhadap resolusi frekuensi dan pengurangan kebocoran.

```python
# Definisi parameter masukan
N = 128  # Panjang sinyal
NFFT = 2048  # Panjang FFT yang digunakan untuk plot spektrum

# Sinyal masukan (panjang N jauh lebih besar dari NFFT)
tt = np.linspace(0, 1, N, endpoint=True)
xx = 100*np.cos(2*np.pi*32*tt) + np.cos(2*np.pi*40*tt)

# Jendela (Kaiser)
wn = sig.kaiser(N, beta=7)
yy = xx * wn  # Sinyal masukan dikali dengan jendela Kaiser

# Hitung FFT 1 (tanpa jendela)
sft = np.abs(fft(xx, NFFT))  # FFT dari sinyal tanpa jendela
slg = 20*np.log10(sft / np.max(sft))  # Konversi menjadi skala logaritmik dalam dB

# Hitung FFT 2 (dengan jendela Kaiser)
yft = np.abs(fft(yy, NFFT))  # FFT dari sinyal dengan jendela Kaiser
ylg = 20*np.log10(yft / np.max(yft))  # Konversi menjadi skala logaritmik dalam dB

# Daftar FFT dan judul jendela untuk plot
lst_ffts = [slg, ylg]
lst_wins = ['No window', 'Kaiser window']

# Plot spektrum sinyal
plt.figure(figsize=(16, 5))
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.title(lst_wins[i])  # Judul plot berdasarkan jenis jendela
    plt.xlabel('freq samples')  # Label sumbu x
    plt.ylabel('level [dB]')  # Label sumbu y
    plt.plot(lst_ffts[i])  # Plot spektrum sinyal
    plt.xlim([0, NFFT//2])  # Batasan sumbu x hingga setengah panjang FFT
    plt.ylim([-80, 0])  # Batasan sumbu y untuk rentang nilai dB
    plt.grid(True)  # Aktifkan grid pada plot
plt.tight_layout()  # Optimalkan tata letak plot
```
